```python
import pandas as pd
import mplfinance as mpl
import matplotlib.pyplot as plt
import os

from pathlib import Path

ROOT_DIR = Path(os.getcwd()).parent
```

```python
df = pd.read_csv(os.path.join(ROOT_DIR, "data/ADAUSDT.csv"))
df = df.convert_dtypes()

df["open_time"] = pd.to_datetime(df["open_time"], unit="ms")
df["close_time"] = pd.to_datetime(df["close_time"], unit="ms")

df = df.sort_values(by="open_time", ascending=True)
```

```python
df_plot = df.copy()
df_plot.index = df_plot["open_time"]

mpl.plot(
    df_plot, type="candle", style="yahoo", warn_too_much_data=int(1e10), title="ADAUSDT"
)
```
